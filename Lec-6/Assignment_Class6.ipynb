1. Stock Market Trend Analysis (Line Plot)  


```python
import numpy as np
import matplotlib.pyplot as plt
days=np.arange(1,31)
prices=np.cumsum(np.random.randn(30)*2+100)
plt.figure(figsize=(10,5),dpi=100)
plt.plot(days,prices,linestyle="-",marker="o",color="b",markersize=6,label="Stock Price")
plt.xlabel("Day")
plt.ylabel("Price (USD)")
plt.title("Stock Price Trend")
plt.legend()
plt.grid(True)
plt.show()
```

2. Customizing Temperature Trends with Line Styles and
Markers

```python
days=np.arange(1,11)
temp_day=np.random.randint(25,35,size=10)
temp_night=np.random.randint(15,25,size=10)
plt.figure(figsize=(8,5))
plt.plot(days, temp_day, linestyle='--', marker='o', color='r',
markersize=8, label="Day Temperature")
plt.plot(days, temp_night, linestyle='-', marker='s', color='b',
markersize=8, label="Night Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation")
plt.legend()
plt.grid()
plt.show()
```


3. Population Growth Visualization (Scatter Plot)

```python
countries = ["USA", "China", "India", "Brazil", "Germany"]
pop_growth=np.array([0.8, 1.2, 1.5, 0.9, 0.5])
population = np.array([331, 1441, 1380, 213, 83])
plt.figure(figsize=(8, 5))
plt.scatter(pop_growth,population,c=population,cmap="viridis",s=population*2,alpha=0.6)
plt.xlabel("Population Growth (%)")
plt.ylabel("Total Population (millions)")
plt.title("Population Growth vs. Total Population")
for i,country in enumerate(countries):
  plt.annotate(country,(pop_growth[i],population[i]),fontsize=10,ha='right')
plt.colorbar(label="Population (millions)")
plt.grid()
plt.show()
```


4. Customer Segmentation with Scatter Plot (Color
Categories)

```python
np.random.seed(42)
age=np.random.randint(18,60,100)
spending_score=np.random.randint(20,100,100)
categories=np.random.choice([1,2,3,4],100)
plt.figure(figsize=(8, 5))
scatter = plt.scatter(age, spending_score, c=categories, cmap='jet',
alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation Analysis")
plt.colorbar(label="Customer Category")
plt.grid()
plt.show()
```
