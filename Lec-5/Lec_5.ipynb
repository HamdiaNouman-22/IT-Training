```python
import numpy as np
A=np.array([[1,2],[3,4]])
B=np.array([[1,2],[3,4]])
print(A@B)
print(np.matmul(A,B))
print(A.T)
print(np.linalg.inv(A))
print(np.linalg.eig(A))
print(np.concatenate((A,B) ,axis=0))
print(np.concatenate((A,B) ,axis=1))
```
